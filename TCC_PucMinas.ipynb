# Adicionando os importes e configurações necessárias

```
import numpy as np 
import pandas as pd

#avoid truncate text
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)
```

# Carregando od datasets

```
# Fonte: https://github.com/datasets/covid-19/tree/master/data
data_time_series = pd.read_csv('/content/drive/My Drive/Colab Notebooks/datasets/time-series-19-covid-combined.csv', encoding = "ISO-8859-1")
display(data_time_series.head())

# Fonte: https://www.kaggle.com/hbfree/covid19formattedweatherjan22march24#covid_dataset.csv
data_weather = pd.read_csv('/content/drive/My Drive/Colab Notebooks/datasets/covid_dataset_weather.csv', encoding = "ISO-8859-1")
display(data_weather.head())

# Fonte: https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)_per_capita
data_gdp_per_capta = pd.read_csv('/content/drive/My Drive/Colab Notebooks/datasets/gdp_nominal_per_capita.csv', encoding = "ISO-8859-1")
display(data_gdp_per_capta.head())
```

# Realizando o tratamento no dataset de infectados do COVID-19

```
# Mostrando como está o dataset
display(data_time_series.describe())

#removendo a coluna [Province/State] que não será utilizada 
data_time_series.drop('Province/State', axis=1, inplace=True)

#Alterando o nome das colunas para padronizar
data_time_series.rename(columns={'Country/Region': 'Country'}, inplace=True)
```

# Realizando o tratamento no dataset de Temperatura

```
# Mostrando como está o dataset
display(data_weather.describe())

#removendo as colunas ['Province/State','smokers','tests','testpop','quarantine','schools','restrictions'] que não serão utilizadas
data_weather.drop(['Province/State','smokers','tests','testpop','quarantine','schools','restrictions'], axis=1, inplace=True)

#Alterando o nome das colunas para padronizar
data_weather.rename(columns={'Country/Region': 'Country', 'lat' : 'Lat', 'long' : 'Long'}, inplace=True)

#Criando a coluna [Date] com as datas a partir da coluna [Day]
from datetime import datetime, timedelta
base_date = datetime(2020, 1, 22)

#Cria o campo Date
data_weather['Date'] = 0

for index, row in data_weather.iterrows():
    new_date = (base_date + timedelta(row['day'])).strftime('%Y-%m-%d')
    data_weather['Date'][index] = new_date
```

# Realizando o tratamento no dataset de Renda Per Capita

```
# Mostrando como está o dataset
display(data_gdp_per_capta.head())

#removendo as colunas [] que não serão utilizadas
#data_gdp_per_capta.drop(['Province/State','smokers','tests','testpop','quarantine','schools','restrictions'], axis=1, inplace=True)

display(data_gdp_per_capta.head())
```

# Concatenando os datasets de Infectados, Temperatura e Renda Per Capita
Para gerar apenas um dataset que será utilizado na solução

```
#display(data_time_series.head(5))
#display(data_weather.head(5))
#display(data_gdp_per_capta.head(5))

#Removendo os espaços
data_gdp_per_capta['Country'] = data_gdp_per_capta['Country'].str.strip()
data_gdp_per_capta['Country'] = data_gdp_per_capta['Country'].str.strip()

# primeiro vamos unir os datasets de infectados com o de renda per capta
result = pd.merge(data_time_series,data_gdp_per_capta, how='inner', on='Country')
display(result.head())
```
