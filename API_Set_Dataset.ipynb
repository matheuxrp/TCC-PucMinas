```
#Cria o dataset de resultado vazio
df_result = pd.DataFrame({'nome': [],'Área': [],'População': [],'Área': [],'Densidade Demográfica': [],'Taxa Geométrica de Crescimento Anual da População - 2010/2020': [],
                          'Grau de Urbanização': [],'Índice de Envelhecimento': [],'População com Menos de 15 Anos': [],'População com 60 Anos e Mais': [],'Razão de Sexos': [],'municipio': []
                          })
df_result.head()
 										
```

```
#display(df_cod_municipio.head(10))

for index,row in df_cod_municipio.iterrows():
  #print(row['cod'])
  response = requests.get("http://api-perfil.seade.gov.br/v1/dados/tema/"+row['cod']+"/all")
  df_data_municipio = response.json()

  #recuperando informações do territorio e populacao
  df_territorio_populacao = pd.DataFrame(df_data_municipio['tema'][0]['dados'])[['nome','mun']]
  df_territorio_populacao.set_index('nome',inplace=True)
  df_territorio_populacao = df_territorio_populacao.T
  #adicionando o nome do municipio
  df_territorio_populacao['municipio'] = df_data_municipio['infolocalidade']['nome']
  #adicionando ao dataframe
  df_result = pd.concat([df_result, df_territorio_populacao])


```
